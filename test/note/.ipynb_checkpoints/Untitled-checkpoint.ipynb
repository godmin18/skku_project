```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#import plotly.graph_objs as go

import os
import warnings
warnings.filterwarnings("ignore")
plt.style.use('ggplot')
```

```python
weather_station_location = pd.read_csv("../data/Weather Station Locations.csv")
weather = pd.read_csv("../data/Summary of Weather.csv")
```

```python
weather_station_location = weather_station_location.loc[:, ["WBAN", "NAME", "STATE/COUNTRY ID", "Latitude", "Longitude"]]
weather = weather.loc[:, ["STA", "Date", "MeanTemp"]]
```

```python
weather_station_location.head(5)
```

```python
weather.head(5)
```

```python
#BINDKURI 지역에 대한 평균 온도
BIN_SID = weather_station_location[weather_station_location.NAME == "BINDUKURI"].WBAN
weather_BIN = weather[weather["STA"] == int(BIN_SID)]
```

```python
weather_BIN["Date"] = pd.to_datetime(weather_BIN["Date"])
```

```python
weather_BIN.head(3)
```

```python
#weather_BIN을 시계열 그래프로 그리기
plt.figure(figsize=(22,8))
plt.plot(weather_BIN.Date, weather_BIN.MeanTemp)
plt.title("Mean Temperature of Bindukuri Area")
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.show()
```

```python
#계절성 주기를 기반으로 시계열 분해
#시계열 분해 참고: https://techblog-history-younghunjo1.tistory.com/81
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(weather_BIN['MeanTemp'], model='additive', freq=7)

fig = plt.figure()
fig = result.plot()
fig.set_size_inches(20,15)
```

```python
#정상성을 판단하기 위한 ACF 그래프
#ACF 값이 천천히 감소하므로 정상성을 만족하지 않는다.
import statsmodels.api as sm

fig = plt.figure(figsize=(50,8))
ax1 = fig.add_subplot(221)
fig = sm.graphics.tsa.plot_acf(weather_BIN['MeanTemp'], lags=20, ax=ax1)
```

```python
# ADF검정으로 정상성 확인
# p-vaule가 0.05를 넘으므로 정상성을 만족하지 못함
from statsmodels.tsa.stattools import adfuller
result = adfuller(weather_BIN['MeanTemp'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Vaules:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```
